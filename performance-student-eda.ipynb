# Preprocessing (categorical variables, Mostly)

```python
Encoded_dataset = pd.get_dummies(dataset,drop_first=True)
Encoded_dataset.head()
```

# Feature Engineering

Creating a Percentage score to the encoded data

```python
Encoded_dataset['percentage'] =  (Encoded_dataset['math score']+Encoded_dataset['reading score']+Encoded_dataset['writing score'])/3
Encoded_dataset.head()
```

Counting No of fails in the Year

```python
Encoded_dataset['Reading Compartments'] = np.where(Encoded_dataset['reading score']>33, 0, 1)
Encoded_dataset['Writing Compartments'] = np.where(Encoded_dataset['writing score']>33, 0, 1)
Encoded_dataset['Math Compartments'] = np.where(Encoded_dataset['math score']>33, 0, 1)
Encoded_dataset['Total Compartments']= (Encoded_dataset['Math Compartments']+Encoded_dataset['Writing Compartments']+Encoded_dataset['Reading Compartments'])
Encoded_dataset.head()
```

```python
Encoded_dataset['Language_percentage'] =  (Encoded_dataset['reading score']+Encoded_dataset['writing score'])/2
Encoded_dataset.head()
```
